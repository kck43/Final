```python
import pandas as pd
import numpy as np
import geopandas as gpd
%matplotlib inline
import matplotlib.pyplot as plt

newdatas = pd.read_csv('datapbl_pgh.csv')
df = pd.DataFrame(newdatas)

neighborhoods = gpd.read_file("Neighborhood_SNAP.shp")

fig, ax=plt.subplots(figsize=(30,30))
neighborhoods.plot(ax=ax, color = "lightcoral")
ax.scatter(df.longitude, df.latitude, c = "black")
plt.show()
```
